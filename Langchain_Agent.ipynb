```python
from langchain.tools import tool
from dotenv import load_dotenv, find_dotenv
from langchain.agents import create_agent
```

```python
env_path = find_dotenv()
load_dotenv(env_path, override=True)
```

```python
agent = create_agent(
    model="gpt-4.1", 
    system_prompt=""" You are a helpful AI assistant. Answer user questions clearly and concisely. """
)

response = agent.invoke( { "messages": [ { "role": "user", "content": "Explain what an AI agent is." } ] } )
response
```

```python
@tool
def get_weather(location: str = "current location") -> str:
    """Get the current weather for a given location."""
    print("🌦️ WEATHER TOOL EXECUTED")

    return f"The current weather in {location} is 20°C and partly cloudy."
```

```python
agent = create_agent(
    model="gpt-4.1", 
    tools=[get_weather]
)


response = agent.invoke( { "messages": [ { "role": "user", "content": "What is today's weather in Tokyo?" } ] } )
response
```
